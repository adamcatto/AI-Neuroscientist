# Plan of Attack
First, I'll use Python's `glob` module to list all files in the specified directory (../../../read_only_sandbox/raw_data). Then, I'll attempt to read each file and determine its format based on its extension. For common spatial transcriptomics formats like CSV, HDF5, or GeoJSON, I'll extract metadata about the dataset (e.g., shape, column names, data types). This will help us understand what kind of data we're working with.

```
import os
import glob
import pandas as pd
import h5py
from nbformat import *


def main():
    # Get all files in the raw_data directory
    file_path = '../../../read_only_sandbox/raw_data'
    files = glob.glob(file_path + '/*')

    # Create a list to store dataset metadata
    datasets = []

    for file in sorted(files):
        filename = os.path.basename(file)
        print(f'Processing file: {filename}')

        try:
            if filename.endswith('.csv') or filename.endswith('.txt'):
                # Read CSV/TXT files
                df = pd.read_csv(file)
                metadata = {
                    'file_name': filename,
                    'type': 'CSV/TXT',
                    'shape': str(df.shape),
                    'columns': list(df.columns)
                }
            elif filename.endswith('.h5') or filename.endswith('.hdf5'):
                # Read HDF5 files
                with h5py.File(file, 'r') as f:
                    metadata = {
                        'file_name': filename,
                        'type': 'HDF5',
                        'keys': list(f.keys())
                    }
            elif filename.endswith('.geojson') or filename.endswith('.json'):
                # Read GeoJSON/JSON files
                with open(file, 'r') as f:
                    data = json.load(f)
                    metadata = {
                        'file_name': filename,
                        'type': 'GEOJSON/JSON',
                        'features_count': len(data.get('features', []))
                    }
            else:
                # Handle unknown file types
                metadata = {
                    'file_name': filename,
                    'type': 'Unknown',
                    'message': f'Unsupported file format: {filename.split('.')[-1]}'
                }

        except Exception as e:
            print(f'Error reading file: {e}')
            metadata = {
                'file_name': filename,
                'type': 'Error',
                'message': str(e)
            }

        datasets.append(metadata)
        print('---')

    return datasets

main()
```
